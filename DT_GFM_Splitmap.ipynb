Create the splitmap

```python
m = leafmap.Map(height=500)
m.add_basemap("Esri.WorldStreetMap")


# GFM Flood in dark blues
gfm_layer = m.add_raster(
    "france/gfm_clean.tif", 
    layer_name="GFM Flood",
    colormap="Blues_r",
    opacity=0.7
)

# DT Extremes in turbo
dt_layer = m.add_raster(
    "france/DTweatherextremes_clean.tif", 
    layer_name="DT Extremes",
    colormap="turbo",
    opacity=0.5
)

# Split control
left_layer = m.layers[2]  # GFM
right_layer = m.layers[3]  # DT  
control = SplitMapControl(left_layer=left_layer, right_layer=right_layer)

m.add(control)
m
```
